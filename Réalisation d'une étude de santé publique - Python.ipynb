## Missions

### Marc

```python
#jointure entre la table population2017 et sous_nutrition2017
pop_ssnutri=pd.merge(population2017,sous_nutrition2017,on ='Zone',how='inner')
pop_ssnutri.head()
```

```python
pop_mondiale=pop_ssnutri['Population'].sum()
prop_ssnutri=pop_ssnutri['nb_sousalim'].sum()
print('La part mondiale de sous nutrition est de ',round(prop_ssnutri/pop_mondiale*100,2),'%')
```

```python
#jointure entre la table population2017 et dispo_alim
dispo_alim=pd.merge(dispo_alim,population2017[['Zone','Population']],on='Zone',how='inner')
dispo_alim.head()
```

```python
dispo_alim['DispoCal']=dispo_alim['Disponibilité alimentaire (Kcal/personne/jour)']*dispo_alim['Population']*365
print("dispo alimentaire totale en kcal :", dispo_alim['DispoCal'].sum(), "kcal")
```

```python
#besoin caloriques d'un être humain : 2500kcal environ
dispo_cal_tot=round(dispo_alim['DispoCal'].sum()/(2500*365))
print ("Nombre total d'être humain pouvant être nourris ",dispo_cal_tot)
proportion=round(dispo_cal_tot/population2017['Population'].sum()*100,2)
print('La proportion est de ',proportion, '%')
```

```python
#mêmes questions avec les végétaux
vegetaux = dispo_alim.loc[dispo_alim['Origine'] == "vegetale"]
vegetaux.head()
```

```python
print("dispo alimentaire totale en kcal des produits végétaux :", vegetaux['DispoCal'].sum(), "kcal")
vegetaux_tot=round(vegetaux['DispoCal'].sum()/(2500*365))
print ("Total d'être humain pouvant être nourrit en végétaux ",vegetaux_tot)
proportion_vegetaux=round(vegetaux_tot/population2017['Population'].sum()*100,2)
print('Proportion : ',proportion_vegetaux,'%')
```

```python
#disponibilité intérieur(animale,pertes,humaines)
proportion_aliments_animaux=round(dispo_alim['Aliments pour animaux'].sum()/dispo_alim['Disponibilité intérieure'].sum()*100,2)
print("La proportion des aliments pour animaux est de " ,proportion_aliments_animaux,"%")
proportion_pertes=round(dispo_alim["Pertes"].sum()/dispo_alim['Disponibilité intérieure'].sum()*100,2)
print("La proportion des pertes est de ",proportion_pertes,"%")
proportion_nourriture=round(dispo_alim["Nourriture"].sum()/dispo_alim['Disponibilité intérieure'].sum()*100,2)
print("La proportion de nourriture utilisé par les humains est de", proportion_nourriture,"%")
```
